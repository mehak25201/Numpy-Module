```python
###BAR CHART
#generate some data
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
```

```python
#Create the plot--barh where h is horizontal
plt.barh(x,y)
plt.show()
```

```python
#ADD labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart Example')
plt.bar(x,y)
plt.show()
```

```python
#JUST EXAMPLE
#complex bar chart
categories=np.array(['A','B','C','D'])
values1=np.array([3,8,1,10])
values2=np.array([6,2,7,5])

#Set the width of the bars
bar_width=0.35
```

```python
#complex group bAR CHART
categories=np.array(['A','B','C','D'])
values1=np.array([3,8,1,10])
values2=np.array([6,2,7,5])
values3=np.array([9,4,8,4])
```

```python
#CREATE PLOT
plt.bar(categories,values1,bar_width,label='Dataset1')
plt.bar(categories,values2,bar_width,label='Dataset2')
plt.bar(categories,values3,bar_width,label='Dataset3')
```

```python
#ADD LABELS AND TITLE
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Group Bar Chart')

#Add legend
plt.legend()


plt.show()
```

```python
#PIE CHART
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
plt.pie(values, labels=categories)
plt.title('Simple pie chart')
plt.show()
```

```python
#to make pie chart effective 
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
#explode highlights the value like exploded the third value by setting 1,2,4=0 and 3=0.1
explode=(0,0,0.1,0)
plt.pie(values, labels=categories,explode=explode)
plt.title('Simple pie chart')
plt.show()
```

```python
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
explode=(0,0,0.1,0)
#autopct show the percentage of category A,B,C,D
#colors=('cyan','pink','purple','blue')
plt.pie(values, labels=categories,explode=explode,autopct='%1.1f%%')
plt.title('Simple pie chart')
plt.show()
```

```python
#Donut chart
#GEnerate data
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
#create plots
fig, ax=plt.subplots()
#Draw the outer ring
ax.pie(values,labels=categories,radius=1,startangle=90,counterclock=False,)
#Draw inner ring
ax.pie([1], radius=0.6,colors='w')
#ADD title
ax.set_title('Donut chart')
plt.show()
```
